```python
import warnings
warnings.filterwarnings("ignore", module="astropy.io.votable.*")
warnings.filterwarnings("ignore", module="pyvo.utils.xml.*")

import io
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit 

# Astropy imports
import astropy.units as u
import astropy.constants as const
from astropy.coordinates import SkyCoord
from astropy.cosmology import Planck15
from astropy.io import votable as apvot

## Generic VO access routines
import pyvo as vo
from uncertainties import ufloat
```

```python
serviceURL="http://tapvizier.u-strasbg.fr/TAPVizieR/tap"
query="""
SELECT z,DL
FROM "J/A+A/557/A12/GRBs"
"""
service=vo.dal.TAPService(serviceURL)
```

```python
resultset = service.search(query)
```

```python
z=[]
D=[]
z_b=[]
D_b=[]
for i in range(len(resultset)):
  if resultset['z'][i]<1:
    z.append(resultset['z'][i])
    D.append(resultset['DL'][i])
  z_b.append(resultset['z'][i])
  D_b.append(resultset['DL'][i])
z=np.array(z)
D=np.array(D)
z_b=np.array(z_b)
D_b=np.array(D_b)
```

```python
def fit_func(x,a):
  return x*a
popt,pcov=curve_fit(fit_func,z,D*10**3)
popt_1,pcov_1=curve_fit(fit_func,z_b,D_b*10**3)
```

```python
fig=plt.figure(figsize=(8,8))
x=np.linspace(0,1,500)
plt.plot(z,D*10**3,'ro',x, popt[0]*x,'b')
plt.xlim((0,1))
plt.ylim((0,7000))
plt.grid()
plt.xlabel(r'$Redshift \hspace{0.5}z$',fontsize=20)
plt.ylabel(r'$D_L \hspace{0.5} Mpc$',fontsize=20)
plt.legend(["Data","Fit"])
plt.show()
```

```python
perr=np.sqrt(np.diag(pcov))
H_0_n=const.c/(popt[0]*u.megaparsec)
H_0_e=perr*u.meter/(u.second*u.megaparsec)
```

$$\text{For} \quad z<1$$
$$H_0=47.485777224 \pm 0.073886152 \frac{\text{km}}{\text{Mpc s}}$$
$$\text{error}=0.15559638\%$$
