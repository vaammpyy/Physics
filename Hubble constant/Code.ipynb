```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress as lr
```

```python
arr=np.loadtxt('hubble_correct.dat')
```

```python
at=arr.T
```

```python
x=at[0]
y=at[1]
```

```python
result=lr(x,y)
```

```python
line= lambda x: result.slope*x+result.intercept
```

```python
fig=plt.figure(figsize=(10,10),facecolor='white')
plt.plot(x*10**-6,y,'ro',x*10**-6,line(x),'k')
plt.xlabel(r'$D\hspace{0.5}[pc]$',fontsize=16)
plt.ylabel(r'$v\hspace{0.5}[km\hspace{0.5} s^{-1}]$',fontsize=16)
plt.title(r'$H_0=%.4e$ $[km\hspace{0.5} s^{-1}\hspace{0.5} Mpc^{-1}]$'%(result.slope*10**6),fontsize=20)
plt.legend(['Data','Fit'])
plt.savefig('Hubb_const.png')
plt.plot()
```
